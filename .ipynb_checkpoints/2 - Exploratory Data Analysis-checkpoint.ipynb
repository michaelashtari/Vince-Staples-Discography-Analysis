As expected, there definitely appears to be some recency bias when it comes to album popularity; his two latest albums have the most recent listens, being the most relevant and timely, driving up their popularity. His first 5 albums, on the other hand, appear to be notably less popular; they all fall anywhere between 14 to 25 points lower than his most recent release!

As far as average track length, interestingly enough, his last 3 releases have averaged the shortest track lengths - between 2-3 minutes per song. ## Valence

To start off, let's take a look at the average valence of his albums. Valence, as defined in Spotify's API documentation, is a continuous variable that ranges from 0 to 1 and measures the "musical positiveness" conveyed by a track. A value closer to 1 indicates more happy and cheerful audio, while a value closer to 0 reflects stronger negative emotions such as anger or sadness.


```python
# Parameters to streamline plot generation
metric = 'valence'
yaxis_max = 0.55
yaxis_min = 0.325

fig, ax = plt.subplots(1, 1, figsize = (10, 10))

# Create line-chart for average valence by album
valence_line = sns.lineplot(data = mean_values_by_album, x = 'album', y = metric, color = 'red', ax = ax)
valence_line.set_title('Average Valence by Album')
valence_line.set_xlabel('Album')
valence_line.set_ylabel('Valence')
valence_line.set_ylim(yaxis_min, yaxis_max)

# Stagger every other x axis label to avoid overlap
for tick in ax.xaxis.get_major_ticks()[0::2]:
    tick.set_pad(20)
    
# Annotate points with album cover and release year
for i, row in mean_values_by_album.iterrows():
    img_response = requests.get(row['album_cover_url']).content
    cover_img = Image.open(BytesIO(img_response))
    cover_offset = OffsetImage(cover_img, zoom = 1)
    
    cover = AnnotationBbox(cover_offset, xy = (i, row[metric] - (yaxis_max - yaxis_min) * .01), frameon = False)
        
    ax.add_artist(cover)
    ax.annotate(row['album_release_date'][0:4], xy = (i, row[metric] + (yaxis_max - yaxis_min) * .05), ha = 'center', va = 'bottom')
    
    
plt.show()
```

Based on the figure above, we can deduce a few things:
* Vince operates on the low to mid scale when it comes to valence; his albums fall within the range of [0.35, 0.55]. 
    * This is unsurprising, as Vince stylistically and thematically leans toward darker topics (mainly concerning the harsh circumstances in which he was raised), and intentionally engineers his sound to coincide with his subject matter. 

* His discography reaches a pronounced global minimum in valence with his ***Big Fish Theory*** album. 
    * This album served as a departure from his typical sound, as it experimented with electro-style production, taking influence from Detroit techno. 
    * Lyrically, Vince uses the intense and percussive sound as a backdrop for his nihilistic lyrics surrounding fame, heartbreak, politics, etc. 
    * His delivery on songs such as "Alyssa Interlude" is very dreary and somber, which emulate a depressive feel; with songs like this, it's no surprise the album clocks in with an average track valence of 0.35.

* His valence global maximum is achieved with his 2018 release, ***FM!*** 
    * This also comes as no shock, as the album is in essence an homage to west coast radio and features some of the liveliest and most energetic deliveries from Vince throughout his entire catalogue. This combines with rapid and vibrant beats to emulate the jubilant summertime experience in southern California.
    * The album is as short as it is animated and attention-grabbing, likely by design, spanning 22 minutes with an average track length of 2 minutes.
    
* Somewhat of a surprising finding is that his ***Prima Donna*** album is not amongst the albums lowest in valence. Nearly each track on the album features a brief spoken-word excerpt from Vince that can be described as raw, emotional, and at times disturbing. These excerpts likely have less of an influence on the overall track's valence score (than I would expect) due to their brevity and lack of production. Being a feature of the overall audio composition of a song, valence is presumably heavily steered by the production/instrumentals across the entirety of the track, rather than the vocalist's speech in a brief outro.

#### In conclusion, the numbers accurately depict Vince's repertoire, and the trajectory captures his intentional experimentation with delivery and production. He tends not to produce "happy"-feeling music as is commonly done in pop music, as his intention is to deliver the honest (and often dystopian) truth behind his experiences, rather than a catchy hook.

## Energy

Next, let's look at the average energy of each album. Per the documentation, energy is yet another continuous numerical variable that ranges from 0 to 1, and measures the intensity and activity of a track. Perceptual features contributing to this attribute include dynamic range, perceived loudness, timbre, onset rate, and general entropy. We can consider death metal tracks as our benchmark of a 1, as they typically feature explosive and intense instrumentals, as well as very perceptually loud and fierce vocals.

Based on this description and given my comments above regarding the valence of notable albums, we should expect to see this feature peak with ***FM!***.

```python
# Parameters to streamline plot generation
metric = 'energy'
yaxis_max = 0.70
yaxis_min = 0.42

fig, ax = plt.subplots(1, 1, figsize = (10, 10))

# Create line-chart for average energy by album
energy_line = sns.lineplot(data = mean_values_by_album, x = 'album', y = metric, color = 'red', ax = ax)
energy_line.set_title('Average Energy by Album')
energy_line.set_xlabel('Album')
energy_line.set_ylabel('Energy')
energy_line.set_ylim(yaxis_min, yaxis_max)

# Stagger every other x axis label to avoid overlap
for tick in ax.xaxis.get_major_ticks()[0::2]:
    tick.set_pad(20)
    
# Annotate points with album cover and release year
for i, row in mean_values_by_album.iterrows():
    img_response = requests.get(row['album_cover_url']).content
    cover_img = Image.open(BytesIO(img_response))
    cover_offset = OffsetImage(cover_img, zoom = 1)
    
    cover = AnnotationBbox(cover_offset, xy = (i, row[metric] - (yaxis_max - yaxis_min) * .01), frameon = False)
        
    ax.add_artist(cover)
    ax.annotate(row['album_release_date'][0:4], xy = (i, row[metric] + (yaxis_max - yaxis_min) * .05), ha = 'center', va = 'bottom')
    
    
plt.show()
```

More deductions!

* Our hypothesis is indeed correct; ***FM!*** is the clear winner when it comes to energy, which can be attributed to the album's audio direction, which masterfully mimicked summertime radio jams.
    * Interestingly, though, it's not ahead of his debut EP, ***Hell Can Wait***, by a large margin. ***Hell Can Wait*** features some rather intense tracks such as the lead-off track "Fire," which makes use of some explosive drum samples that somewhat resemble the type of energy one might expect from a metal track.

* Vince's albums tend to hover somewhere around the 0.60 mark, with the very clear outlier being his self-titled ***Vince Staples*** album, clocking in at about 0.46.

    * The ***Vince Staples*** album is quantified as the least energetic in his catalogue likely because, stylistically, it presents Vince in perhaps the most vulnerable manner that he's conveyed in any of his music. Many tracks feature more acoustic elements in the instrumental, which serves as a backdrop to his relatively soft-spoken rapping while he delves into a number of personal topics, all of which translates to a grounded deviation from his regular performances. Vince utilizes this low-energy, melancholy approach to emphasize the intimacy of these subjects.

* Consistent with our observations from the Valence section, ***Prima Donna*** continues to sit with the norm when it comes to energy, despite its frequent somber and monotone spoken word sections. This speaks to the superior weight of the actual musical component of the song when it comes to determining these metrics.

It's important to keep in mind that while the former is the most acoustic of the bunch, it's still relatively low on the measure's scale, which runs all the way up to 1. 
    * This makes sense, as the former album contains a higher proportion of songs that have prominent acoustic *samples* than any other album, but isn't a true acoustic album intrinsically, as it still contains much electrical amplification (as is characteristic of hip-hop in general).

* Interestingly, Vince's debut EP ***Hell Can Wait*** is the minimum this time around, which isn't too shocking given that the production on the album features a lot of interesting samples and sounds, few of which resemble real musical instruments in isolation. I am surprised to see that ***Big Fish Theory*** scores marginally higher, as the latter is a heavily electronic-influenced project, which would lead me to believe that it has close to no acoustic properties.

## Tempo

Next, let's gauge the tempo of each album, which is a musical term that describes the speed or pace of the beat in a song. Tempo is powerful in that manipulating it can drastically impact the mood of a song, as a very quick song is likely more amenable to dancing, movement, and energy, whereas a slower tempo may invite more negative emotions. 

In this case, tempo is a numerical variable measured in beats per minute (BPM), and thus can take on any value >= 0, although genres tend to have defining ranges.

```python
# Parameters to streamline plot generation
metric = 'tempo'
yaxis_max = 135
yaxis_min = 100

fig, ax = plt.subplots(1, 1, figsize = (10, 10))

# Create a line-chart for average tempo by album
tempo_line = sns.lineplot(data = mean_values_by_album, x = 'album', y = metric, color = 'red', ax = ax)
tempo_line.set_title('Average Tempo by Album')
tempo_line.set_xlabel('Album')
tempo_line.set_ylabel('Tempo')
tempo_line.set_ylim(yaxis_min, yaxis_max)

# Stagger every other x axis label to avoid overlap
for tick in ax.xaxis.get_major_ticks()[0::2]:
    tick.set_pad(20)
    
# Annotate points with album cover and release year
for i, row in mean_values_by_album.iterrows():
    img_response = requests.get(row['album_cover_url']).content
    cover_img = Image.open(BytesIO(img_response))
    cover_offset = OffsetImage(cover_img, zoom = 1)
    
    cover = AnnotationBbox(cover_offset, xy = (i, row[metric] - (yaxis_max - yaxis_min) * .01), frameon = False)
        
    ax.add_artist(cover)
    ax.annotate(row['album_release_date'][0:4], xy = (i, row[metric] + (yaxis_max - yaxis_min) * .05), ha = 'center', va = 'bottom')
    
    
plt.show()
```

As someone who isn't very well-versed in music theory, these results are particularly profound to me. A quick [Google search](https://learningmusic.ableton.com/make-beats/tempo-and-genre.html) provided the following tempo ranges for common genres:

* Hip-Hop: 60-100 bpm
* House: 115-130 bpm
* Techno/trance: 120-140 bpm
* Dubstep: 135-145 bpm

Given this benchmark and the plotted results, it appears that the pace of Vince's music is generally closer to House than it is to Hip-Hop, which is interesting, as I might've expected this for albums like ***Summertime '06*** and ***Big Fish Theory***, but not for his entire discography.

Other notable findings:

* ***Prima Donna*** and ***Vince Staples*** have the highest average tempo, at roughly 128 bpm. This is surprising, since these two albums make no effort to provide a dancy and energetic experience, yet appear to be high-paced and even etch into the "Techno" genre given the benchmark ranges above. The latter is arguably the furthest from electronic in his discography (as evidenced by it being the acousticness apex), and yet shockingly matches the pace of the genre.

    * This is further evidence that genres are multifactored, and moreover that the impact/emotion evoked by a track is a complex product of its audio features; just because ***Vince Staples*** employed a higher tempo resembling a techno pace, doesn't mean it is in effect a dance album. The reference ranges above are not 1-to-1 with the genres they appear next to, but simply reflect common choices amongst artists of that genre. 

* To my immense shock, ***Big Fish Theory***, an album with techno-inspired instrumentals, is not amongst the albums with highest tempo. 
    * Although the tracks in BFT have many layers to their production, they're not necessarily high-paced. This is yet again more reinforcement of our first finding. There's much to be said about the complexity of the instrumentals in the album in terms of the variety of sounds relayed at any given moment, but the pacing is not substantially different from the other records. 
    * Apparently this is a common confusion with tracks that exhibit very "active" production - they seem like they might be higher paced due to all of the concurrent musical activity, but the underlying beat may still be following a relatively low frequency.

* The fact that ***RAMONA PARK BROKE MY HEART*** takes on the minimum tempo of the set does make sense. The album as a whole has a much slower and more digestible feel to it than Vince's previous releases. Songs such as "WHEN SPARKS FLY" feature production that is reminiscient of R&B slow jams. As a whole, the album feels remarkably different than its predecessors, and evidently tempo has a large part to do with that.

Turns out I wasn't completely on the mark with my predictions.

* First of all, it's clear that Vince's albums sport a wide range of danceability. 
    * ***Hell Can Wait*** is understandably perceived to be relatively less danceable than the others at 0.56, while ***FM!*** is highly danceable at 0.83. These two extremes make sense intuitively given the very different thematic directions of the two albums, which also correspond with the sonic differences between the two. 
        * ***Hell Can Wait*** is described as a grimy reflection of Vince's mindset in his youth, in which he could do what he wanted and figure out (or neglect) the consequences later. 
        * ***FM!*** served as a critique of the popular consumption of contemporary black art, masquerading as a terrestrial radio show. 
        * The former took on grimier production to support its inauspicious nature, while the latter sonically achieved its goal of providing a mainstream and accessible, and thus danceable, sound through its use of bass-heavy and lively instrumentals.
* The positive tempo-danceability correlation I predicted does not appear to be prominently at work here. 
    * Both ***Prima Donna*** and ***Vince Staples*** appear to be around average of the pack, despite being tempo leads. 
    * From this data, it seems that danceability is more strongly swayed by the deeper properties of the beat, such as the energy and perceived loudness, rather than the pace, hence why ***FM*** towers amongst the others despite it being bang on average when it comes to pacing.
    * I am however surprised that ***Big Fish Theory*** doesn't deviate further from the norm, considering the qualities of its production seem much more "dancy" than the others. This expectation stemmed from the fact that the album exhibits the most techno-like properties, which I've come to associate with dance music, but that evidently is not the strongest inference.
    
* Generally speaking, Vince has converged on a fairly danceable style of music (averaging around 0.70) over the trajectory of his career since his debut. The sheer difference between the min and max of this value speaks to his versatility as an artist - it's clear that not every album of his is designed to have the listener moving.

* A caveat to point out here (which also applies to virtually all other audio features) is that these album averages are computed across all songs in the album, *unweighted* by duration or any other variable. 
    * This is potentially problematic, as a non-musical interlude track that lasts about 30 seconds may have substantial impact on the average metric of an album, especially in albums with fewer tracks. 
    * For example, ***FM!*** is a particularly energetic album, yet has two interludes (of 11 total tracks) that score low in energy due to the fact that they feature little instrumental, and focus more on spoken word. These outliers have noticeable impact on the album's average energy, since they assume 18% weight in the unweighted average, thus slightly skewing the calculation. 
    * In regards to danceability, interludes are generally designed and implemented in order to add to and support the overall *theme* of an album, rather than necessarily be appreciated/danced to as musical tracks. Giving interludes equal strength over the average danceability as any other track will inevitably punish concept albums that feature many such interludes. Such is the drawback of taking on an unweighted approach when computing these album-level metrics.

# Intra-Album Comparisons

Up to this point we've looked at the tracks in Vince Staples' discography at a bird's eye view (via our high-level descriptive statistics), as well as studied them aggregately within their album groupings (inter-album comparisons), as collections of tracks related by a signature theme or sound. ## Caveats and Limitations

Naturally, however, there are some caveats to our approaches that may have skewed some findings, and thus are paramount to acknowledge (as is the case in any analysis). Some of these include:

* Sample size concerns
    * As aforementioned, Vince's discography consists of only 83 tracks, which makes for a pretty small dataset. 
    * As the law of large numbers would insist, using statistical means to generalize a relatively small sample can lead to easily skewed results at the hand of outliers. Working with a larger dataset would allow us to more comfortably and reliably recognize trends in the data and draw conjecture from it.
    * This issue takes on an even more pronounced role when we further the group the small dataset into albums, which in the case of Vince, can sometimes take on as few as 6 tracks or as many as 20!
    
* Album-level metric computations
    * As pointed out toward the end of the *Danceability* section, our album averages were computed across all tracks in the album, with each track having equal weight. 
    * This can be problematic when dealing with a collection of tracks that have high varability in song length, and especially when skits/spoken-word tracks/interludes are included in the mix. 
        * Interludes have dramatically different sound structure from musical tracks, and thus their audio metrics look quite different - not to mention, they tend to be much shorter as well. Giving these tracks equal strength over an album average can heavily skew those averages, particularly for concept albums that are dense in these sorts of interludes!
    * An alternative approach would be to compute album averages using track length (in seconds) as weights. This way, the tracks that make up a larger proportion of the listening experience gain adequate representation over the album's metric, while the shorter and more fleeting tracks have less influence (and aren't discarded entirely). It would be worth reperforming this analysis under this approach to see how results differ.
    
* Audio features are limited to conveying audio structure
    * It's important to recognize the limitations of using audio features to characterize musical albums. In lyrical music, ESPECIALLY in hip-hop, audio and production do play a strong role in contriving an experience and conveying a message, but they only take on half of the show. The other equally as important (and in some cases, arguably more important) component is the actual lyrical content itself.
    * While much of the commentary in this notebook alluded to theme and lyrical content, this analysis did little to actually quantify, study, and analyze Vince's lyrics themselves. Our feature analysis primarily dissected the production and delivery of tracks and albums with the intention of recognizing signature traits of albums' sound structure, and then matching those styles to the already familiar lyrical/thematic contexts.
